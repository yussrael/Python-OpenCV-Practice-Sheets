Q2 load the following image and apply sharpen filter.

Observation: Using Kernel sharpening matrix (edges are minus ones and inner is 9) results the image with a quite sharper quality

```python

def SHOW(img):
    display(Image.fromarray(img));

image = cv2.imread('C:\\Users\\Lenovo\\Desktop\\blurred.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#image = img[:,:,::-1]


kernel_sharpen_1 = np.array([[-1, -1, -1],
                             [-1,  9, -1],
                             [-1, -1, -1]])

sharp_image_1 = cv2.filter2D(image, -1, kernel_sharpen_1)

f, (ax0,ax1) = plt.subplots(1, 2, figsize=(25,15))
ax0.imshow(image),ax0.set_title('blurred')
ax1.imshow(sharp_image_1),ax1.set_title('Sharp')

```

Q2 load the following image and apply blur filter 


Observation: the image goes from very sharp to smooth and blurry, using two different ways in each image blurred_1 and blurred 2
(avereging and kernel 5x5 matrix with values of 1/25)

```python
import cv2
img = cv2.imread('C:\\Users\\Lenovo\\Desktop\\sharpen.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel1 = np.ones((5,5),np.float32)/25
dst1 = cv2.filter2D(img, -1, kernel1, borderType=cv2.BORDER_ISOLATED)

kernel2= cv2.blur(img,(9,9))

f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(30,15))
ax0.imshow(img),ax0.set_title('sharp_original')
ax1.imshow(dst1),ax1.set_title('blurred_1')
ax2.imshow(kernel2),ax2.set_title('blurred_2')
```
